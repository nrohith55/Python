## Matplotlib
Matplotlib is a plotting library for the Python programming language and its numerical mathematics extension NumPy.
Benifits of Matplotlib library are as follows

Generally easy to get started for simple plots
Support for custom labels and texts
Great control of every element in a figure
High-quality output in many formats
Very customizable in general

```python
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Sine wave")
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.cos(x)
plt.plot(x,y)
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Cosine wave")
plt.show()
```

```python
x=np.arange(0,4*np.pi,0.1) # To plot sine and cosine waves togethef
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x,y_sin,"bo")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Sine wave")

plt.subplot(2,1,2)
plt.plot(x,y_cos,"g*")
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Cosine wave")

plt.show()
```

#Seaborn Library
Distribution plots
Below plots are mainly used to represent Numeric values/features in the dataset

1) Dist Plot 2) Join Plot 3) Pair Plot

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
df.head()
```

```python
#Correlation and Heat Map
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
# Pairplot : It is basically a scatter plot used in finding the correlation between the variables
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
# Dist Plot
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10) # kde =Kernel density estimation
```

## Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

boxplot
violinplot
countplot
bar plot

```python
sns.countplot('sex',data=df) # Countplot
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
df.head()
```

```python
sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x="day",y="total_bill",data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x="total_bill",y="day",hue='smoker',data=df)# categorize the data based on some other categories
```

```python
sns.violinplot(x="total_bill",y='day',data=df,palette='rainbow')
#Violin plot helps us to see both the distribution of data in terms of Kernel density estimation and the box plot
```
